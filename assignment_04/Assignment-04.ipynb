Open census_cost.txt into line_list and print one line at a time using a loop

```
with open("census_cost.txt", "rt") as f:
    line_list = f.readlines()
for row in line_list:
    print(row)
```

Extract the first two lines of line_list and print them

```
top2_list=line_list[:2]
for row in top2_list:
    print(row)
```

Extract the data from line_list and print

```
data_list=line_list[2:]
for row in data_list:
    print(row)
```

```
#Split each row in data_list with '\t' 
#This will result in four sublists of data_list, one for each column
for i in range(0, len(data_list)):
    data_list[i]=data_list[i].split('\t')
```

```
#Test to ensure the first element of the first row is what is expected
data_list[0][0]
```

Extract the column Census Year, clean, and display 

```
#Create an empty list for the first column
year_list=[]
for i in range(0, len(data_list)):
    #Append each year to year_list
    year_list.append(data_list[i][0])
    #If there is an extra character (*) in the year, delete it
    if len(year_list[i])>4:
        year_list[i]=year_list[i].replace('*','')
for row in year_list:
    print(row)
```

Extract the column Total Population, clean, and display 

```
#Create an empty list for the second column
pop_list=[]
for i in range(0, len(data_list)):
    #Append each population count to pop_list
    pop_list.append(data_list[i][1])
    #Remove the ',' in each number
    pop_list[i]=pop_list[i].replace(',','')
for row in pop_list:
    print(row)
```

Extract the column Census Cost, clean, and display

```
#Create an empty list for the third column
cost_list=[]
for i in range(0, len(data_list)):
    #Append each Census Cost to cost_list
    cost_list.append(data_list[i][2])
    #Remove the ',' and '$' from each number
    cost_list[i]=cost_list[i].replace(',','')
    cost_list[i]=cost_list[i].replace('$','')
    if 'Billion' in cost_list[i]:
        cost_list[i]=cost_list[i].replace('Billion','')
        cost_list[i]=str(int(float(cost_list[i])*1000000000))
        
#Remove 'Billion' and add '0's to those values 
#cost_list[21]='4500000000'
#cost_list[22]='13000000000'
for row in cost_list: 
    print(row)
```

Extract the column Average Cost Per Person, clean, and display

```
#Create an empty list for the forth column
avg_list=[]
for i in range(0, len(data_list)):
    #Append the Avg Cost data to avg_list
    avg_list.append(data_list[i][3])
    #If the value is in cents, remove 'cents', and divide number by 100 to display value in dollars 
    if 'cents' in avg_list[i]:
        avg_list[i]=avg_list[i].replace(' cents','')
        avg_list[i]=str(round(float(avg_list[i])/100,2))
    #If the value is in dollars, remove the '$'
    if '$' in avg_list[i]:
        avg_list[i]=avg_list[i].replace('$','')
    #Remove the '\n' from each item in list
    avg_list[i]=avg_list[i].replace('\n','')
for row in avg_list:
    print(row)
```

Consolidate the new data to a csv file 

```
#Open a new csv file to put cleansed data
with open("census_cost.csv", "wt") as f:
    #While writing data to the new file, replace '\t' with ',' in the top two lines
    for line in top2_list[0:]:
        f.write(line.replace("\t", ","))
    #While writing data to the new file, add a comma after each column item, and a new line after the last column 
    for line in range(0, len(year_list)):
        f.write(year_list[line]+',')
        f.write(pop_list[line]+',')
        f.write(cost_list[line]+',')
        f.write(avg_list[line]+'\n')
```
