```
# Create a consolidated df
df_c=df[['UNITID', 'INSTNM', 'CITY', 'STABBR', 'ZIP', 'ADM_RATE', 'UGDS', 'TUITIONFEE_IN', 'MN_EARN_WNE_P10']]
```

# Colsolidated Dataframe Variable Definitions

UNITID = Unit ID of Insitution

INSTNM = Insitution name

CITY = City

STABBR = State

ZIP = ZIP Code

ADM_RATE = Admission Rate

UDGS = Enrollment of undergraduate certificate/degree-seeking students

TUITIONFEE_IN = In-state tuition and fees

MN_EARN_WNE_P10 = Mean earnings of students working and not enrolled 10 years after entry
