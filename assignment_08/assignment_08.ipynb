Based on the differences calculated when using one oberservation vs. 20 observations, we can theorize that the more samples we use in our calculations, the closer the calculated mean will be to the mean of the whole set. 

```
#17
#Write a function to input sample size and data, calculate means, and return difference
def mean_diff(size, df):
  s_list=[]
  for x in range(0, size):
    randnum=random.randint(0,len(df)-1)
    s=df['TUITIONFEE_IN'][randnum]
    s_list.append(s)
  m_sample=round(sum(s_list)/len(s_list),2)
  m_total=round(df.TUITIONFEE_IN.mean(),2)
  diff=round(abs(m_total-m_sample),2)
  return(diff)
```

```
#18
#Test function
mean_diff(2000, df)
```

```
#19
sample_sizes=list(range(1,51))
sample_sizes
```

```
#20
#Use the function from block #17 (mean_diff), using each sample size from the sample_sizes list
means_diff=[]
for x in sample_sizes:
  means_diff.append(mean_diff(x, df))
means_diff
```

```
#21
#Plot the results from sample_sizes
fig, ax=plt.subplots(figsize=(12,8))
#Create plot
sns.scatterplot(ax=ax, x=sample_sizes, y=means_diff).set_title('Sample Mean and Total Mean Difference (50 samples)')
#Edit plot 
plt.xlabel("Sample Size")
plt.ylabel("Sample Mean - Total Mean")
```

```
#Repeat the process with 301 samples
sample_sizes=list(range(1,301))
means_diff=[]
for x in sample_sizes:
  means_diff.append(mean_diff(x, df))
#Create Plot
fig, ax=plt.subplots(figsize=(12,8))
sns.scatterplot(ax=ax, x=sample_sizes, y=means_diff).set_title('Sample Mean and Total Mean Difference (300 samples)')
#Edit Plot 
plt.xlabel("Sample Size")
plt.ylabel("Sample Mean - Total Mean")
```
