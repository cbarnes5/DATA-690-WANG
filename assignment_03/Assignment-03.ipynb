## Print 100 random integers between 0 and 9 in a 10 x 10 matrix

```
import random
# Generate 10 rows of 10 random integers between 0 and 9
for i in range(0, 10):
    for j in range(0, 10):
        value = random.randint(0, 9)
        # end=" " provides a space after each printed number
        print(value, end=" ")
    # Start a new line
    print("")
```

## If the random number is an odd number, print an @ instead 

```
import random
# Generate 10 rows of 10 random integers between 0 and 9
for i in range(0, 10):
    for j in range(0, 10):
        value = random.randint(0, 9)
        # If the random value generated is not divisable by 2 with no remanders, value=@
        if value % 2 != 0:
            value="@"
        # end=" " provides a space after each printed number
        print(value, end=" ")
    # Start a new line
    print("")
```

## Calculate and print the total of each row 

```
import random
# Create a list
h_list=[]
# Generate 10 rows of 10 random integers between 0 and 9
for i in range(0, 10):
    # Create a sublist inside h_list for each row
    h_list.append([])
    for j in range(0, 10):
        value = random.randint(0, 9)
        # Append value to sublist
        h_list[i].append(value)
        print(value, end=" ")
        # Sum all the values in the sublist
        listsum=sum(h_list[i])
    # Print the listsum after a * once the inside loop is completed (i.e. a row is completed)
    print("*", listsum)
```

## Surround the matrix with asterisks

```
import random
# Print top line of *
print("*" * 23)
# Generate 10 rows of 10 random integers between 0 and 9
for i in range(0, 10):
    for j in range(0, 10):
        value = random.randint(0, 9)
        # If this is the first value generated for each line, print a * first
        if j==0:
            print("*", value, end=" ")
        else: 
            print(value, end=" ")
    # Print the last *
    print("*")
# Print the bottom line of *
print("*" * 23)
```

## Calculate and print the total of each row and column 

```
import random
# Create a list
h_list=[]
# Generate 10 rows of 10 random integers between 0 and 9
for i in range(0, 10):
    # Create a list inside a list for each row
    h_list.append([])
    for j in range(0, 10):
        value = random.randint(0, 9)
        # Append value to h_list
        h_list[i].append(value)
        print(value, end="  ")
        # Sum all the values in h_list
        listsum=sum(h_list[i])
    # Print the listsum after a *
    print("*", listsum)
# Print an * under each column 
print("*  " * 10)
# Create a list
v_list=[]
# Start a loop to focus on each column of the matrix
for j in range(0,10):
    # Create a list inside a list for each column
    v_list.append([])
    # For each row in the matrix, put the value of the column of interest into v list
    for i in h_list:
        v_list[j].append(i[j])
    # Print the sum of each sublist in v_list under the correct column 
    print(sum(v_list[j]), end=" ")
```
